#### Approximating $f(x) = x^3$ on the $[−3, 2]$ interval

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1) Function to define a neural architecture

```python
from pybrain.structure import LinearLayer
from pybrain.structure import TanhLayer
from pybrain.structure import FeedForwardNetwork
from pybrain.structure import FullConnection
```

```python
def build_network(hidden_dim):
    net = FeedForwardNetwork()
    inLayer = LinearLayer(1)
    hiddenLayer = TanhLayer(hidden_dim)
    outLayer = LinearLayer(1)
    
    net.addInputModule(inLayer)
    net.addModule(hiddenLayer)
    net.addOutputModule(outLayer)
    
    
    in_to_hidden = FullConnection(inLayer, hiddenLayer)
    hidden_to_out = FullConnection(hiddenLayer, outLayer)
    
    net.addConnection(in_to_hidden)
    net.addConnection(hidden_to_out)
    net.sortModules()
    
    return net
```

```python
from pybrain.datasets import SupervisedDataSet
from numpy.random import uniform
def f(x):
    return x**3

def build_dataset(low, high, n=1000):
    ds = SupervisedDataSet(1,1)
    values=uniform(low,high, n)
    for i in values:
        ds.addSample(i, f(i))
    return ds
    
```

2) Preparing datasets

```python
train=build_dataset(-3, 2, 2000)
test1=build_dataset(-3, 2, 1000)
test2=build_dataset(-3, 3, 1000)
```

```python
#[−3, 1] ∪ [2, 3].
train_2=SupervisedDataSet(1,1)
values=list(uniform(-3,1,1000))+list(uniform(2,3,1000))
for i in values:
    train_2.addSample(i, f(i))
```

```python
#Function to plot the mean squared error
from pybrain.tools.validation import ModuleValidator, Validator
from pybrain.supervised import BackpropTrainer

def plt_errors(train):
    dim=[i for i in range(1,100,10)]
    errors1=[]
    errors2=[]
    for hd in dim:
        net=build_network(hd)
        trainer = BackpropTrainer(net,weightdecay=1e-05,dataset=train)
        trainer.trainEpochs(100)
        err1 = ModuleValidator.validate(Validator.MSE, trainer.module, test1)
        err2 = ModuleValidator.validate(Validator.MSE, trainer.module, test2) 
        errors1.append(err1)
        errors2.append(err2)
    plt.plot(dim,errors1)
    plt.plot(dim,errors2)
    plt.legend(['Initial interval','Extended interval'],bbox_to_anchor=(1.2, 1))
    plt.xlabel('hidden layer dimension')
    plt.ylabel('MSE')
    plt.title('MSE from hidden layer dimensionality')
    plt.show()
    
dim=[1, 5, 10, 20, 50, 100, 200]
test=np.sort(uniform(-3,3,1000))
def plt_approx(train):
    results=[]
    legend=[]
    for hd in dim:
        print hd
        net=build_network(hd)
        trainer = BackpropTrainer(net,weightdecay=1e-05,dataset=train)
        trainer.trainEpochs(100)
        pred = [net.activate(i)[0] for i in test[:, None]]
        results.append(pred)
        legend.append('%d neurons'%hd)
    results.append([f(x) for x in test])
    legend.append('$x^{3}$')
    plt.figure(figsize=(12,10))
    for i in range(len(dim)+1):
        plt.plot(test, results[i])  
    plt.legend(legend,bbox_to_anchor=(1.2, 1))
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('$x^{3}$ approximation for varying number of neurons')
    plt.show()
```

3) Ploting MSE

```python
plt_errors(train)
```

The minimum number of neurons in hidden layer enough to approximate the given function on [−3, 2] interval is 40. Low values of layer dimension underfit while there is a sharp rise of MSE for 50 which indicates overfitting

MSE on extended interval is slightly higher because the weights of the network were fit for the given interval only.

The MSE values are higher for small numbers of neurons because of their poor generalization ability. More neurons provide nearly the same result as for the continuous interval and the errors for both intervals (initial and extended) сonverge

Large numbers of neurons for this training interval cause overfitting. Nevertheless, 100 neuronss seem to be optimal for both train datasets

5) This function transfroms an input image into a set of activations of the hidden layer neurons. 

```python
def activate_on_hidden(net, ds):
    X=[]
    y=[]    
    for i in range(ds.getLength()):
        net.activate(ds.getSample(i)[0])
        target=ds.getSample(i)[1]
        y.append(list(target).index(1))
        X.append(np.array(net['hidden'].outputbuffer[net['hidden'].offset]))  
    return X,y
```

```python
mnist_train_hidden_X, mnist_train_hidden_y=activate_on_hidden(net, mnist_train)
mnist_test_hidden_X, mnist_test_hidden_y=activate_on_hidden(net, mnist_test)
```

```python
total_X=np.r_[mnist_train_hidden_X,mnist_test_hidden_X]
total_y=np.r_[mnist_train_hidden_y,mnist_test_hidden_y]
```
